```python
benua15 = whr_15.groupby('Region')['Happiness Score'].mean()

fig = plt.figure(figsize=(25,6))
ax = fig.add_subplot(111)

ax.bar(x=benua15.index, height=benua15.values)
ax.set_title('Happiness Score 2015 Based on Region')
ax.set_xlabel('Region')
ax.set_ylabel('Happiness Score')
plt.xticks(rotation=60)
plt.show()
```

```python
top10 = whr_15[['Country', 'Happiness Score']].sort_values(by='Happiness Score', ascending=False).head(10)

fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111)

ax.bar(x=top10['Country'], height=top10['Happiness Score'])
ax.set_title('Top 10 Countries Based on Happiness Score 2015')
ax.set_xlabel('Countries')
ax.set_ylabel('Happiness Score')
```

All the 10 countries are developed countries and most of them are known to have very high median income which relate to high GDP and strong economy. Looking at this, we have a hypothesis: GDP has high correlation with happiness score.

Happiness scores for top 10 countries are above 7 with insignificant difference. What's interesting is Australia and New Zealand, which has the highest score based on region (shown in previous viz), are at the bottom compared to other 8 countries. Is it because Western Europe and Northern America happiness score have wider range? Let's see the discrepancy in each region

```python
whr_15.groupby('Region')['Happiness Score'].max()-whr_15.groupby('Region')['Happiness Score'].min()
```

```python
beda15 = whr_15.groupby('Region')['Happiness Score'].max()-whr_15.groupby('Region')['Happiness Score'].min()

fig = plt.figure(figsize=(15,6))
ax = fig.add_subplot(111)

ax.bar(x=beda15.sort_values().index, height=beda15.sort_values().values, )
ax.set_title('Difference of Happiness Score 2015 in each Region')
ax.set_xlabel('Region')
ax.set_ylabel('Happiness Score Difference')
plt.xticks(rotation=25)
plt.show()
```

As expected, region Australia and New Zealand has the lowest gap (nearly 0). North America has a slight gap (second lowest gap across all region), whereas Western Europe is among the top 3 who has quite significant gap value. This might answer why Ausralia and New Zealand ranked lowest in the top 10 highest happiness score chart.

```python
semua = whr_19[['Overall rank', 'Country or region', 'Score', 'GDP per capita','Social support', 'Healthy life expectancy',
       'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']].sort_values(by='Score', ascending=False).head(95)

fig = plt.figure(figsize=(28,6))
ax = fig.add_subplot(111)
semua.plot(x='Country or region',y=['GDP per capita','Social support', 'Healthy life expectancy','Freedom to make life choices', 'Generosity', 
'Perceptions of corruption'], kind='bar', stacked=True, ax=ax)
plt.title('Countries Happiness Score Broken Down into Attributes (2019)', fontdict={'size': 16})
plt.style.use('seaborn-pastel')
plt.legend(loc='upper right')
plt.show()
```

Although the data has been sorted based on rank (score), apparently, the stacked bar chart (which consists of all 6 attributes beside score and rank) showed inconsistency. If we see it closer, it turns out that the happiness scores in 2019 do not equal to sum of all the 6 available attributes. We are not gonna change the data (of course), but it's interesting to see an anomaly in Singapore and Hong Kong. These 2 countries show high score in all 6 attributes but they do not rank on top. Looking at this anomaly (which also happens in other countries like Costa Rica and Mexico), it seeems that the other variable(s) that do not included in the dataset has quite an effect on the rank. Nevertheless, it's only a guess, we should not conclude anything for now because of data limitation.
